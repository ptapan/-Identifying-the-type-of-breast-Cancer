# Tuning epochs epchos and batch size

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
 def create_model():
        
    # define model
        model = keras.Sequential([
        keras.Input(shape=X_train.shape[1]),
        Dense(64, activation="relu"),
        Dense(32, activation="relu"),
        Dense(16, activation="relu"),
        Dense(1,activation='sigmoid')
         ])
     
    # compile model
        model.compile(loss='binary_crossentropy', metrics=['accuracy'])
        return model
    
    
seed = 7
tf.random.set_seed(seed)   
model = KerasClassifier(build_fn=create_model)
# define the grid search parameters
batch_size = [8,16, 32, 64, 128, 256]
epochs = [ 100,120,140,150,180,200]
#optimizer = ['sgd', 'rmsprop', 'adam']
#earning_rate = [0.001, 0.01, 0.1, 0.2, 0.3]
parameter = dict(batch_size=batch_size, epochs=epochs)
rnd_cv1 = RandomizedSearchCV(estimator=model, param_distributions=parameter, n_jobs=-1, cv=3)
rnd_cv1.fit(X_train, y_train)
rnd_cv1.best_params_
```

```python
# Tuing optimization Learnig_rate
```

```python
 def create_model(learning_rate=3e-3):
        
    # define model
        model = keras.Sequential([
        keras.Input(shape=X_train.shape[1]),
        Dense(64, activation="relu"),
        Dense(32, activation="relu"),
        Dense(16, activation="relu"),
        Dense(1,activation='sigmoid')
         ])
        optimizer = keras.optimizers.Adam(lr=learning_rate)
    # compile model
        model.compile(loss='binary_crossentropy', metrics=['accuracy'],optimizer=optimizer)
        return model
    
    
seed = 7
tf.random.set_seed(seed)   
model = KerasClassifier(build_fn=create_model)
# define the grid search parameters
learning_rate = [0.001, 0.01, 0.1, 0.2, 0.3]
parameter = dict(learning_rate=learning_rate)
rnd_cv1 = RandomizedSearchCV(estimator=model, param_distributions=parameter, n_jobs=-1, cv=3)
rnd_cv1.fit(X_train, y_train,epochs=100,batch_size=16)
rnd_cv1.best_params_
```

# Building a Model With Tuned parameter

```python
def create_model(learning_rate=1e-3):
        
    # define model
        model = keras.Sequential([
        keras.Input(shape=X_train.shape[1]),
        Dense(64, activation="relu"),
        Dense(32, activation="relu"),
        Dense(16, activation="relu"),
        Dense(1,activation='sigmoid')
         ])
        optimizer = keras.optimizers.Adam(lr=learning_rate)
    # compile model
        model.compile(loss='binary_crossentropy', metrics=['accuracy'],optimizer=optimizer)
        return model
    
    
seed = 7
tf.random.set_seed(seed)   
model = KerasClassifier(build_fn=create_model)
model.fit(X_train,y_train,epochs=140,batch_size=16)
```

```python
prediction = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,roc_auc_score,auc,roc_curve
plt.figure(figsize=(12,10))
sns.heatmap(confusion_matrix(y_test,prediction), annot=True,cmap='coolwarm')
```

```python
print(classification_report(y_test,prediction))
```

# AUC

```python
auc = roc_auc_score(y_test, prediction)
auc
```

# ROC

```python
fpr, tpr, thresholds = roc_curve(y_test, prediction)
```

```python
plt.figure(figsize=(15,10))
sns.set_style('darkgrid')
plt.plot(fpr, tpr, color='orange', label='ROC')
plt.plot([0, 1], [0, 1], color='darkblue', linestyle='--',label='ROC curve (area = %0.2f)' % auc)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
plt.show()
```
